```python
# Import necessary libraries

import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score

from xgboost import XGBClassifier

from sklearn.impute import SimpleImputer

from sklearn.preprocessing import LabelEncoder

from sklearn.datasets import load_iris

 
# Load the Iris dataset

iris = load_iris()

data = pd.DataFrame(data=iris.data, columns=iris.feature_names)

data["class"] = iris.target_names[iris.target]

 

# Display the first few rows of the dataset

print("Original Dataset:")

print(data.head())
```

```python
data.isna().sum()
```

```python
data.describe([i/100 for i in range(100)])
```

```python
data.rename(columns = {'class':'target'}, inplace = True)
```

```python
data.target.unique()
```

```python
data
```

```python
y = data['target']
X = data.drop('target',axis=1)
```

```python
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, train_size = .75)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
y_train.head(10)
```

```python
y_train
```

```python
label_encoder = LabelEncoder()

y_train_encode = label_encoder.fit_transform(y_train)
y_test_encode = label_encoder.fit_transform(y_test)

print('y_train', y_train_encode)
print('y_test', y_test_encode)
```

```python
label_encoder.classes_
```

```python
print(y_train_encode[0],y_train.reset_index(drop = True)[0])
# print(y_train_encode[2],y_train[2])
```

```python
import seaborn as sns

# Concatenate features and target variable for visualization
data_visualization = pd.concat([X, pd.Series(y, name='target')], axis=1)

# Create a pair plot
sns.pairplot(data_visualization, hue='target', markers=["o", "s", "D"], palette="Set2")
```

```python
xgb_classifier = XGBClassifier(base_score=0.5, colsample_bylevel=1,
       colsample_bytree=1, gamma=0, learning_rate=0.1, max_delta_step=0,
       max_depth=3, min_child_weight=1, n_estimators=100,
       n_jobs=1, nthread=None, objective='binary:logistic', random_state=0,
       reg_alpha=0, reg_lambda=1, subsample=1)
xgb_classifier
```

```python
xgb_classifier.fit(X_train, y_train_encode)
```

```python
y_preds_train = xgb_classifier.predict(X_train)
```

```python
y_preds_train
```

```python
y_train_encode
```

```python
y_pred_test = xgb_classifier.predict_proba(X_test)
```

```python
y_pred_test
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Train the model on the training data
xgb_classifier.fit(X_train, y_train_encode)

# Make predictions on the training data
y_train_pred = xgb_classifier.predict(X_train)

# Make predictions on the testing data
y_test_pred = xgb_classifier.predict(X_test)

# Calculate accuracy for training and testing data
accuracy_train = accuracy_score(y_train_encode, y_train_pred)
accuracy_test = accuracy_score(y_test_encode, y_test_pred)

# Display the accuracy
print("Accuracy on Training Data:", accuracy_train)
print("Accuracy on Testing Data:", accuracy_test)

# Classification Report for Testing Data
print("\nClassification Report for Testing Data:")
print(classification_report(y_test_encode, y_test_pred))

# Confusion Matrix for Testing Data
print("\nConfusion Matrix for Testing Data:")
print(confusion_matrix(y_test_encode, y_test_pred))
```
